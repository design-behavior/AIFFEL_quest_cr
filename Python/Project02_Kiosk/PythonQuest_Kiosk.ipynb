```python
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]

class Kiosk:

    def __init__(self):
        self.menu = menu
        self.price = price
        self.order_menu = [] # self.order_menu 에 대한 초기화 설정이 없어 오류 발생 > 초기화 설정으로 해결
        self.order_price = [] # self.order_price 에 대한 초기화 설정이 없어 오류 발생 > 초기화 설정으로 해결
        self.price_sum = 0 # self.price_sum 에 대한 초기화 설정이 없어 오류 발생 > 초기화 설정으로 해결

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):

        print()  # 줄 바꿈

        n = 0
        while n < 1 or len(menu) < n: # 주문가능한 숫자가 될 때까지 물어보기
            n = int(input("음료의 번호를 입력하세요 : "))  # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n <= len(menu):
                self.price_sum = self.price[n-1]  # 선택 음료의 가격
                print(self.menu[n-1], ' : ', self.price[n-1], '원')  # 메뉴

            # 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 음료 온도 물어보기
        t = 0  # 기본값을 넣어주고
        temp = "" # 메소드가 temp 속성을 가지고 있지 않아 오류 발생 > self.temp > temp로 수정하고 temp = "" 추가
        while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

            # HOT 음료를 선택했을 때
            if t == 1:
                temp = 'HOT' # 메소드가 temp 속성을 가지고 있지 않아 오류 발생 > self.temp > temp로 수정하고 temp = "" 추가

            # ICE 음료를 선택했을 때
            elif t == 2:
                temp = 'ICE' # 메소드가 temp 속성을 가지고 있지 않아 오류 발생 > self.temp > temp로 수정하고 temp = "" 추가

            else:
                print("1과 2 중 하나를 입력하세요.\n")

        self.order_price.append(self.price[n-1])  # 가격 리스트에 추가
        self.order_menu.append(temp + ' ' + self.menu[n-1])  # 주문 리스트에 추가
        self.price_sum += self.price[n-1] # 추가 주문 금액 더하기 (합계금액)

        print('주문 음료', temp, self.menu[n-1], ' : ', self.price[n-1], '원')  # 온도 속성을 추가한 주문 결과 출력

        print()  # 줄 바꿈

        while n != 0: # 지불(0)을 선택하기 전까지 물어보기

            print()  # 줄 바꿈

            n = int(input("추가 주문은 음료 번호, 결제는 0을 입력하세요 : "))  # 추가 주문 또는 결제 선택
            if 1 <= n <= len(menu):

                print()  # 줄 바꿈

                # 추가 음료 온도 물어보기
                t = 0
                while t != 1 and t != 2: # 1이나 2를 입력할 때까지 물어보기
                    t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

                    # HOT 음료를 선택했을 때
                    if t == 1:
                        temp = 'HOT' # 메소드가 temp 속성을 가지고 있지 않아 오류 발생 > self.temp > temp로 수정하고 temp = "" 추가

                    # ICE 음료를 선택했을 때
                    elif t == 2:
                        temp = 'ICE' # 메소드가 temp 속성을 가지고 있지 않아 오류 발생 > self.temp > temp로 수정하고 temp = "" 추가

                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                self.order_price.append(self.price[n-1])  # 가격 리스트에 추가
                self.order_menu.append(temp + ' ' + self.menu[n-1])  # 주문 리스트에 추가
                self.price_sum += self.price[n-1] # 추가 주문 금액 더하기 (합계금액)

                print('주문 음료', temp, self.menu[n-1], ' : ', self.price[n-1], '원', '지금까지 합계 : ', self.price_sum, '원')  # 온도 속성을 추가한 주문 결과 출력

            elif n == 0: # 지불(0)을 선택하기 전까지 물어보기

                print()  # 줄 바꿈

                print("주문이 완료되었습니다.")
                print(self.order_menu, self.order_price, '총 주문금액 : ', self.price_sum)

            else:

                print()  # 줄 바꿈

                print("없는 메뉴입니다. 다시 주문해 주세요.")

    # 결제 메서드
    def pay(self):

        p = 0
        while p != 1 and p != 2:  # 1이나 2를 입력할 때까지 물어보기
            p = int(input("현금결제는 1을, 카드결제는 2를 입력하세요 : "))  # 결제방법 번호 입력

            # 현금결제를 선택했을 때
            if p == 1:
                print('현금결제 > 직원을 호출하겠습니다.')  # 현금결제 선택 시 메시지 출력

            # 카드결제를 선택했을 때
            elif p == 2:
                print('카드결제 > IC칩 방향에 맞게 카드를 꽂아주세요.')  # 카드결제 선택 시 메시지 출력

            else:
                print("1과 2 중 하나를 입력하세요.\n")

    # 주문서 출력 메서드
    def table(self):
        # 외곽
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 상품명 : 해당 금액
        for menu, price in zip(self.order_menu, self.order_price): # for i in range 형태로 작성한 것에 오류 > self.order_menu와 self.order_price를 zip에 변수를 2개로 나누어 각각 출력하여 해결
            print(menu, ' : ', price, '원')

        print('합계 금액 :', self.price_sum, '원')

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')

a = Kiosk()
a.menu_print()
a.menu_select()
a.pay()
a.table()

```
