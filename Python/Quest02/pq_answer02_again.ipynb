```python
# 복사한 경로
# /content/drive/MyDrive/Colab Notebooks/pq_answer/Avengers.txt

f = open("/content/drive/MyDrive/Colab Notebooks/pq_answer/Avengers.txt", 'r')
data = f.read()

import string

# data.lower().replace(c, "")를 이용하여 소문자로 정리하고 특수문자를 삭제한다.
for c in string.punctuation:
  data = data.lower().replace(c, "")

# .split()을 이용하여 띄워쓰기 단위로 자른다.
split_data = data.split()
print("소문자로 교체, 띄워쓰기 단위로 자른 리스트 : ", split_data)

print("-" * 100)

# for 문을 이용하여 비어 있는 딕셔내리에 단어와 단어수를 key:vaule를 넣는다. (알듯말듯한 코드)
word_counts = {}
for word in split_data:
  if word in word_counts:
   word_counts[word] += 1
  else:
   word_counts[word] = 1
print("단어별 빈도수 카운팅 : ", word_counts)

print("-" * 100)

# sorted를 이용하여 내림차순으로 정리한다. (모르는 코드)
sorted_dict = sorted(word_counts.items(), key = lambda item:item[1], reverse=True)
print("빈도수 높은 순으로 리스트 정렬 : ", sorted_dict)

print("-" * 100)

# dict를 이용하여 정렬된 (키, 값) 쌍의 리스트를 딕셔너리로 변환한다.
dicted_dict = dict(sorted_dict)
print("빈도수 높은 순으로 딕셔너리 정렬 : ", dicted_dict)

print("-" * 100)

# unzip하여 딕셔내리를 튜플로 분리한다.
word_ranking, counting = zip(*sorted_dict)
print("unzip한 단어순 : ", word_ranking)
print("unzip한 빈도수순 : ", counting)

print("-" * 100)

# unzip한 단어순에 정수 인덱스를 부여한다. (알듯말듯한 코드)
word_ranking_numbering=[]
for i in enumerate(word_ranking):
  word_ranking_numbering.append(i)
dictionary=dict(word_ranking_numbering)
print("정수 인덱스 부여 : ", dictionary)

# unzip한 단어순에 정수 인덱스 부여한다. 2 (알듯말듯한 코드)
word_ranking_numbering = list(enumerate(word_ranking, start=1))  # 인덱스 1부터 시작
dictionary = dict(word_ranking_numbering)
print("정수 인덱스 부여 2 : ", dictionary)

print("-" * 100)

text=input("english only : ")
t_text = text.split()
result = []
for t in t_text:
  result_rank = word_ranking.index(t)+1 #딕셔너리는 인덱스 속성이 없어서 딕셔너리(word_ranking_numbering)이 아닌, 리스트(word_ranking)를 대입해야 하는 듯
  result.append(result_rank)
print(result)
```
