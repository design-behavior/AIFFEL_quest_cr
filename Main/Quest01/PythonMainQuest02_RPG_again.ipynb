```python
import random # random ë¶ˆëŸ¬ì˜¤ê¸°

class Character:

    def __init__(self, name, level, energy, attack_p, defense_p):
        self.name = name
        self.level = level
        self.energy = energy
        self.attack_p = attack_p
        self.defense_p = defense_p

    def is_alive(self):
        return self.energy > 0 # return bool(self.energy >= 0)ë¥¼ ìˆ˜ì •

    def take_damage(self, damage):
        if damage > self.defense_p:
            damage_inflicted = damage - self.defense_p
            self.energy -= damage_inflicted

        else:
            damage_inflicted = 0

        if self.energy < 0: # ë¬¸ë²• ì˜¤ë¥˜ ìˆ˜ì • (ì½œë¡  ì¶”ê°€)
            self.energy = 0

        return damage_inflicted

    def attack_target(self):
        # damage ë³€ìˆ˜ ì‚­ì œ (ë°ë¯¸ì§€ëŠ” 1ë¶€í„° ê³µê²©ë ¥ ì‚¬ì´ì˜ ëœë¤í•œ ì •ìˆ˜)
        return random.randint(1, self.attack_p)


class Player(Character):

    def __init__(self, name, level=1, energy=100, attack_p=25, defense_p=5, exp=0): # ê²½í—˜ì¹˜ ì†ì„±(exp) ì¶”ê°€
        super().__init__(name, level, energy, attack_p, defense_p) # class Character ë¶ˆëŸ¬ì˜¤ê¸°, # self ì†ì„± ì‚­ì œ (super()ê°€ ìë™ìœ¼ë¡œ selfë¥¼ ì „ë‹¬í•¨ìœ¼ë¡œ)
        # self.level = 1 (ë¶€ëª¨ í´ë˜ìŠ¤ì™€ì˜ ì¼ê´€ì„± ìœ ì§€ë¥¼ ìœ„í•´ ì¬ì„ ì–¸ ì‚­ì œ)
        # self.energy = 100 (ë¶€ëª¨ í´ë˜ìŠ¤ì™€ì˜ ì¼ê´€ì„± ìœ ì§€ë¥¼ ìœ„í•´ ì¬ì„ ì–¸ ì‚­ì œ)
        # self.attack_p = 25 (ë¶€ëª¨ í´ë˜ìŠ¤ì™€ì˜ ì¼ê´€ì„± ìœ ì§€ë¥¼ ìœ„í•´ ì¬ì„ ì–¸ ì‚­ì œ)
        # self.defense_p = 5 (ë¶€ëª¨ í´ë˜ìŠ¤ì™€ì˜ ì¼ê´€ì„± ìœ ì§€ë¥¼ ìœ„í•´ ì¬ì„ ì–¸ ì‚­ì œ)
        self.exp = exp

    def gain_experience(self, exp_points):
        self.exp += exp_points
        print(f"{self.name} gained {exp_points} EXP. Total EXP: {self.exp}")  # ê²½í—˜ì¹˜ ì¦ê°€ ë©”ì‹œì§€ ì¶”ê°€ (ì´ ë¶€ë¶„ ì¡°ê¸ˆ ë” ì‚´í´ë³´ê¸°)
        self.level_up()

    def level_up(self):
        while self.exp >=50: # if -> whileë¡œ ë³€ê²½í•˜ì—¬ ì—¬ëŸ¬ ë²ˆ ë ˆë²¨ ì—… ê°€ëŠ¥í•˜ê²Œ ìˆ˜ì •
            self.exp -= 50  # ê²½í—˜ì¹˜ ì°¨ê° ì¶”ê°€ (ê²½í—˜ì¹˜ ì°¨ê°ì´ ì—†ìœ¼ë‹ˆ, ë¬´í•œë°˜ë³µí•¨)
            self.level += 1
            self.attack_p += 10
            self.defense_p += 5
            print(f"{self.name} leveled up to Level {self.level}! "
                  f"Attack: {self.attack_p}, Defense: {self.defense_p}")  # ë ˆë²¨ ì—… ë©”ì‹œì§€ ì¶”ê°€ (ì´ ë¶€ë¶„ ì¡°ê¸ˆ ë” ì‚´í´ë³´ê¸°)

class Monster(Character):

    def __init__(self, name, level): # energy, attack_p, defense_pì˜ ì‚­ì œí•˜ê³  í•˜ë‹¨ì— ë³„ë„ ì •ì˜
        energy = random.randint(10, 30) * level # ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ë©”ì†Œë“œë¥¼ í™œìš©í•˜ì—¬ ì´ˆê¸°í™”ë¥¼ í•œë²ˆì— ì²˜ë¦¬í•˜ëŠ” ê²ƒì´ ì¼ê´€ë˜ê³  íš¨ìœ¨ì ì´ë¼ê³  í•¨
        attack_p = random.randint(5, 20) * level # ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ë©”ì†Œë“œë¥¼ í™œìš©í•˜ì—¬ ì´ˆê¸°í™”ë¥¼ í•œë²ˆì— ì²˜ë¦¬í•˜ëŠ” ê²ƒì´ ì¼ê´€ë˜ê³  íš¨ìœ¨ì ì´ë¼ê³  í•¨
        defense_p = random.randint(1, 5) * level # ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ë©”ì†Œë“œë¥¼ í™œìš©í•˜ì—¬ ì´ˆê¸°í™”ë¥¼ í•œë²ˆì— ì²˜ë¦¬í•˜ëŠ” ê²ƒì´ ì¼ê´€ë˜ê³  íš¨ìœ¨ì ì´ë¼ê³  í•¨

        super().__init__(name, level, energy, attack_p, defense_p) # class Character ë¶ˆëŸ¬ì˜¤ê¸°



def battle(player, monster):
    """ Playerì™€ Monsterê°€ ì‹¸ìš°ëŠ” ì „íˆ¬ í•¨ìˆ˜ """
    print(f"\nâš”ï¸ {player.name} vs {monster.name} (Lv. {monster.level}) ì „íˆ¬ ì‹œì‘!")

    while player.is_alive() and monster.is_alive():
        # Playerì˜ ê³µê²©
        player_damage = player.attack_target()
        damage_dealt = monster.take_damage(player_damage)
        print(f"{player.name}ì˜ ê³µê²©! {monster.name}ì—ê²Œ {damage_dealt} í”¼í•´ë¥¼ ì…í˜”ë‹¤! (ë‚¨ì€ ì²´ë ¥: {monster.energy})")

        if not monster.is_alive():
            print(f"{monster.name}ì„(ë¥¼) ì²˜ì¹˜í–ˆë‹¤!")
            player.gain_experience(monster.level * 20)
            print("ğŸ‰ ì „íˆ¬ ìŠ¹ë¦¬!")
            return True  # ìŠ¹ë¦¬

        # Monsterì˜ ê³µê²©
        monster_damage = monster.attack_target()
        damage_received = player.take_damage(monster_damage)
        print(f"{monster.name}ì˜ ê³µê²©! {player.name}ì´(ê°€) {damage_received} í”¼í•´ë¥¼ ì…ì—ˆë‹¤! (ë‚¨ì€ ì²´ë ¥: {player.energy})")

        if not player.is_alive():
            print("â˜ ï¸ ì „íˆ¬ íŒ¨ë°°..")
            return False  # íŒ¨ë°°


def main():
    """ ê²Œì„ ì‹¤í–‰ í•¨ìˆ˜ """
    monster_dict = {
        'ìŠ¬ë¼ì„': 1,
        'ê³ ë¸”ë¦°': 2,
        'ì˜¤í¬': 3
    }

    # Player ìƒì„±
    player_name = input("í”Œë ˆì´ì–´ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ")
    player = Player(player_name)

    # ëª¬ìŠ¤í„°ì™€ ì°¨ë¡€ë¡œ ì „íˆ¬ ì§„í–‰
    for monster_name, monster_level in monster_dict.items():
        monster = Monster(monster_name, monster_level)

        if not battle(player, monster):
            print("\nğŸ’€ ê²Œì„ ì˜¤ë²„! ë” ì´ìƒ ì „íˆ¬ë¥¼ ì§„í–‰í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return  # í”Œë ˆì´ì–´ê°€ íŒ¨ë°°í•˜ë©´ ê²Œì„ ì¢…ë£Œ

    print("\nğŸŠ ëª¨ë“  ëª¬ìŠ¤í„°ë¥¼ ì²˜ì¹˜í–ˆìŠµë‹ˆë‹¤! ê²Œì„ í´ë¦¬ì–´!")


# ê²Œì„ ì‹¤í–‰
if __name__ == "__main__":
    main()
```
